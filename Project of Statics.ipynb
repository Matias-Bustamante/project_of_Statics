```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
file="Ventas+Videojuegos.xlsx"
data=pd.read_excel(file, sheet_name="Ventas Videojuegos") #Importamos el archivo xlsx 
```

```python
data_año=data.groupby('Año')[['Ventas NA',"Ventas EU", "Ventas JP"]].sum() #Calculamos la cantidad de ventas por año en NA, EU, JP
data_año.head(20)
```

```python
#Crear Graficos 

def graficos(data, columns, fig, posicion, text): 
    
    ax=fig.add_subplot(3,1,posicion)
    
    ax.bar(data.index, columns)
    ax.set_title(text.capitalize(), color='Red', fontsize=30)
    ax.set_xlabel("AÑO", color='blue', fontsize=30)
    ax.set_ylabel(text.capitalize(), color="blue", fontsize=20)
    plt.tick_params(labelsize = 30)
    


fig=plt.figure(figsize=(40,30))

graficos(data_año, data_año['Ventas NA'], fig, 1, 'Ventas NA') #Ventas NA por Año
graficos(data_año, data_año['Ventas EU'], fig, 2, "Ventas EU") #Ventas EU por año 
graficos(data_año, data_año['Ventas JP'],fig, 3, "Ventas JP") #Ventas JP por año 
plt.show()
```

```python
#Calculamos la Media de Ventas de NA, EU, JP
media_ventas_na=data['Ventas NA'].mean()
media_ventas_eu=data['Ventas EU'].mean()
media_ventas_jp=data['Ventas JP'].mean()
```

```python
print('Media NA:',media_ventas_na)
print('Media EU', media_ventas_eu)
print('Media JP: ', media_ventas_jp)
```

```python
std_ventas_na=data['Ventas NA'].std()
std_ventas_eu=data['Ventas EU'].std()
std_ventas_jp=data['Ventas JP'].std()

print('Desvio NA:',std_ventas_na)
print('Desvio EU', std_ventas_eu)
print('Desvio JP: ', std_ventas_jp)
```
